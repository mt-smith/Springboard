## Loading & Cleaning Dataset

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
takehome_users = pd.read_csv('takehome_users.csv', encoding= 'latin-1')
takehome_engagement = pd.read_csv('takehome_user_engagement.csv', encoding = 'latin-1')
```

```python
print(takehome_users.shape)
takehome_users.creation_time = pd.to_datetime(takehome_users.creation_time)
takehome_users.last_session_creation_time = pd.to_datetime(takehome_users.last_session_creation_time, unit='s')
takehome_users = takehome_users.rename(columns={"object_id": "user_id"})
takehome_users.invited_by_user_id = takehome_users.invited_by_user_id.fillna(0).astype(int)
takehome_users.info()
```

```python
takehome_users.head()
```

```python
#Creation of feature for account age when user last logged in
takehome_users['account_age_at_most_recent_login'] = (
    takehome_users['last_session_creation_time'] - takehome_users['creation_time']
    )\
    .dt.days
#Creation of feature for current account age
takehome_users['current_account_age'] = (max(takehome_users['creation_time']) - takehome_users['creation_time'])\
    .dt.days
```

```python
print(takehome_engagement.shape)
takehome_engagement.time_stamp = pd.to_datetime(takehome_engagement.time_stamp)
takehome_engagement.info()
```

```python
takehome_engagement.head()
```

```python
def get_rolling_login(grp, freq):
    return grp.rolling(freq, on='time_stamp')['visited'].sum()

takehome_engagement['rolling_login_counter'] = takehome_engagement.groupby('user_id', as_index=False, group_keys=False) \
                            .apply(get_rolling_login, '7D')

takehome_engagement['rolling_login_counter'].value_counts()
```

```python
takehome_engagement['adopted_user'] = np.where((takehome_engagement['rolling_login_counter'] >= 3), \
                                               1, 0)
```

```python
active_filter = takehome_engagement[takehome_engagement.adopted_user == 1]
active_filter = active_filter.drop_duplicates(subset=['user_id'])
active_filter
```

```python
df = takehome_users.merge(active_filter[['user_id', 'adopted_user']], how ='left', on='user_id')
df.adopted_user = df.adopted_user.fillna(0).astype(int)
df.last_session_creation_time = df.last_session_creation_time.fillna(df['creation_time'])
df.head()
```

## Summary

* Account age at most recent login is the most relevant feature when predicting whether a user will convert to an "adopted user". There is a sharp decline in "account age at most recent login" for non-adopted users. This indicates that as a user continues logging in further from their account creation date, they are more likely to become an adopted user. 

Relax can leverage this insight to devleop features that make users more likely to use the product over time, such as notifications prompting user to login.
